```python
#      NumPy
import numpy
mylist = [[1, 2, 3], [3, 4, 5]]
myarray = numpy.array(mylist)
print(myarray)
print(myarray.shape)
print(" : %s" % myarray[0])
print(" : %s" % myarray[-1])
print("      : %s" % myarray[0, 2])
print(" : %s" % myarray[:, 2])
```
