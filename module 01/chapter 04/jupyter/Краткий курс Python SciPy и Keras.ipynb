### Работа со списками
Списки используют нотацию в квадратных скобках и могут индексироваться с помощью адресации, подобной массивам. 
Обратите внимание, что мы используем некоторые простые функции форматированного вывода для объединения строк и переменных при печати.

```python
# Работа со списком
mylist = [1, 2, 3]
print("Нулевой элемент: %d" % mylist[0])
mylist.append(4)
print("Длина списка: %d" % len(mylist))
for value in mylist:
    print(value)
```

### Работа со словарями
Словари – это сопоставления имен со значениями, например пары ключ-значение. Обратите внимание на использование фигурных скобок и двоеточий при определении словаря.

```python
# Работа со словарем
mydict = {'a': 1, 'b': 2, 'c': 3}
print("Значение для A: %d" % mydict['a'])
print("Значение для B: %d" % mydict['b'])
mydict['a'] = 11
print("Значение для A: %d" % mydict['a'])
print("Ключи: %s" % mydict.keys())
print("Значения: %s" % mydict.values())
for key in mydict.keys():
    print(mydict[key])
```

### Определение функций
В приведенном ниже примере используется новая функция для вычисления суммы двух значений и вызывается функция с двумя аргументами.

## Краткий курс NumPy
NumPy предоставляет программисту базовые структуры данных и операции для SciPy. Это массивы (ndarrays), которые можно эффективно определять и которыми удобно манипулировать.
Перед началом использования библиотеки в программе ее необходимо импортировать

```python
# Импорт библиотеки
import numpy
# Определение массива NumPy
mylist = [1, 2, 3]
myarray = numpy.array(mylist)
print(myarray)
print(myarray.shape)
```

Обратите внимание, как мы легко преобразовали список Python в массив NumPy:

```python
# Доступ к значениям массива в NumPy
import numpy
mylist = [[1, 2, 3], [3, 4, 5]]
myarray = numpy.array(mylist)
print(myarray)
print(myarray.shape)
print("Первая строка: %s" % myarray[0])
print("Последняя строка: %s" % myarray[-1])
print("Элемент на пересечении заданных строки и столбца: %s" % myarray[0, 2])
print("Целый столбец: %s" % myarray[:, 2])
```

Массивы NumPy можно использовать непосредственно в записи арифметических операций

```python
# Арифметические операции над массивами NunPy
import numpy
myarray1 = numpy.array([2, 2, 2])
myarray2 = numpy.array([3, 3, 3])
print("Поэлементное сложение массивов %s" % (myarray1 + myarray2))
print("Поэлементное перемножение массивов : %s" % (myarray1 * myarray2))
```

# Краткий курс Matplotlib
Библиотеку Matplotlib можно использовать для создания диаграмм и графиков. Самые частые виды использования данной библиотеки следующие:
- Вызов функции построения графиков с некоторыми данными (например, .plot() ).
- Вызов множества функций для настройки внешнего вида графика (например, этикетки – labels и цвета – colors).
- Переключение графика в видимый режим (например, .show() ).

В приведенном ниже примере кода Python создается простой линейный график из одномерных данных. 


```python
# Простой линейный график
import matplotlib.pyplot as plt
import numpy
myarray = numpy.array([1, 2.5, 3])
plt.plot(myarray)
plt.xlabel('Ось X')
plt.ylabel('Ось У')
plt.show()
```

Код ниже реализует простой пример создания точечной диаграммы из двухмерных данных. 

```python
# Простой пример точечного графика
import matplotlib.pyplot as plt
import numpy
x = numpy.array([1, 2, 3])
y = numpy.array([2, 5, 6])
plt.scatter(x,y)
plt.xlabel('Ось X')
plt.ylabel('Ось У')
plt.show()
```

## Краткий курс Pandas
Pandas предоставляет структуры данных и функциональные возможности для быстрого манипулирования данными и их анализа. Ключом к пониманию применения Pandas в задачах машинного обучения, включая работу с нейронными сетями, которая без данной библиотеки вообще невозможна, является понимание структур данных Series и DataFrame.
### Серии (Series) в Python
Ряд (серию) в Python можно представить как одномерный массив, в котором строки и столбцы могут быть маркированы, помечены особыми метками.

```python
# Работа с сериями Pandas
import numpy
import pandas
myarray = numpy.array([1, 2, 3])
rownames = ['a', 'b', 'c']
myseries = pandas.Series(myarray, index=rownames)
print(myseries)
```

Следует отметить, что в действительности **Series** – **не массив**, а **динамическая структура**, как и образованный на ее основе **Dataframe**, который мы будем рассматривать в следующем пункте.
К элементу серии можно обратиться и как к элементу массива NumPy, и как к элементу словаря (см. код ниже). 

```python
print(myseries[0])
print(myseries['a'])
```

Dataframe или фрейм данных можно представить как многомерный массив, в котором строки и столбцы могут быть сопоставлены с метками.

```python
# Работа с датафреймами в Pandas
import numpy
import pandas
myarray = numpy.array([[1, 2, 3], [4, 5, 6]])
rownames = ['a', 'b']
colnames = ['раз', 'два', 'три']
mydataframe = pandas.DataFrame(myarray, index=rownames, columns=colnames)
print(mydataframe)
```

Данные можно индексировать с помощью имен столбцов.

```python
print("Первый метод:")
print("Столбец-раз: %s" % mydataframe['раз'])
print("Второй метод:")
print("Столбец-раз: %s" % mydataframe.раз)
```

Обратите внимание, что индексация действует и при использовании кириллических наименований столбцов, что не может не радовать.
