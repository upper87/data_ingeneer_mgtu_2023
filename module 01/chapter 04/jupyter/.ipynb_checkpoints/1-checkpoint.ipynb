Синхронизация с БД постгриСКЛ создание бд по домашке

```python
from configparser import ConfigParser
from typing import Dict
import psycopg2

def load_connection_info(ini_filename : str) -> Dict[str, str]:
    parser = ConfigParser()
    parser.read(ini_filename)
    conn_info = {param[0]: param[1] for param in parser.items("postgresql")}
    return conn_info

def create_db(conn_info : Dict[str, str]) -> None:
    psql_connection_string = f"host={conn_info['host']} port={conn_info['port']} user={conn_info['user']} password={conn_info['password']}"
    conn = psycopg2.connect(psql_connection_string)
    cur = conn.cursor()
    conn.autocommit = True
    sql_query = f"CREATE DATABASE hw_parkinson"
#    sql_query = f"CREATE DATABASE {conn_info['database']}"
    
    try:
        cur.execute(sql_query)
    except Exception as e:
        print(f"{type(e).__name__}: {e}")
        print(f"Query: {cur.query}")
        cur.close()
    else:
        conn.autocommit = False
        
def create_table(sql_query :str,
                conn: psycopg2.extensions.connection,
                cur: psycopg2.extensions.cursor) -> None:
    try:
        cur.execute(sql_query)
    except Exception as e:
        print(f"{type(e).__name__}: {e}")
        print(f"Query: {cur.query}")
        conn.rollback()
        cur.close()
    else:
        conn.commit()
        
        

conn_info = load_connection_info("db.ini")


print(conn_info)
#Создание бд - закоментил т.к. уже создана бд
#create_db(conn_info)

connection = psycopg2.connect(**conn_info)
cursor = connection.cursor()

sql_add_table1 = """
CREATE TABLE IF NOT EXISTS homework.sound
(
    id_patient integer NOT NULL,jitter_local numeric NOT NULL,jitter_local_abs numeric NOT NULL,jitter_rap numeric NOT NULL,jitter_ppq5 numeric NOT NULL,jitter_ddp numeric NOT NULL,flicker_local numeric NOT NULL,flicker_local_db numeric NOT NULL,flicker_apq3 numeric NOT NULL,flicker_apq5 numeric NOT NULL,flicker_apq11 numeric NOT NULL,flicker_dda numeric NOT NULL,"AC" numeric NOT NULL,"NTH,HTN" numeric NOT NULL,median_tone numeric NOT NULL,middle_tone numeric NOT NULL,std_tone numeric NOT NULL,min_tone numeric NOT NULL,max_tone numeric NOT NULL,puls_count numeric NOT NULL,period_count numeric NOT NULL,middle_period numeric NOT NULL,std_period numeric NOT NULL,share_of_unvoiced_fragments numeric NOT NULL,number_of_voice_breaks numeric NOT NULL,degree_brokenness_speech numeric NOT NULL,"UPDRS" numeric NOT NULL,disease integer NOT NULL DEFAULT 1,
    CONSTRAINT pat_id FOREIGN KEY (id_patient)
        REFERENCES homework.person (sound_id_patient) MATCH SIMPLE
        ON UPDATE CASCADE
        ON DELETE CASCADE
        NOT VALID
)
"""

sql_add_table2 = """
CREATE TABLE IF NOT EXISTS homework.person
(
    sound_id_patient integer NOT NULL,
    id_global integer NOT NULL GENERATED ALWAYS AS IDENTITY ( INCREMENT 1 START 1 MINVALUE 1 MAXVALUE 2147483647 CACHE 1 ),
    name "char" NOT NULL,
    CONSTRAINT person_pkey PRIMARY KEY (sound_id_patient)
)
"""
sql_add_table3 = """
CREATE TABLE IF NOT EXISTS homework.image
(
    "X_axis" numeric NOT NULL,y_axis numeric NOT NULL,z_axis numeric NOT NULL,pressure numeric NOT NULL,"сapture_angle" numeric NOT NULL,time_label numeric NOT NULL,test_id integer NOT NULL,
    CONSTRAINT image_pkey PRIMARY KEY (test_id)
)
"""

#СОздание схемы создал и закоментил
create_table("CREATE SCHEMA IF NOT EXISTS homework", connection, cursor)
create_table(sql_add_table1, connection, cursor)
create_table(sql_add_table2, connection, cursor)
create_table(sql_add_table3, connection, cursor)


connection.close()
cursor.close()


```

Заполнение данными наших таблиц из целевых файлов 

```python
import pandas as pd
import psycopg2
import psycopg2.extras as psql_extras
from typing import Dict
from configparser import ConfigParser
from typing import Dict, List

def load_connection_info(ini_filename: str) -> Dict[str, str]:
    parser = ConfigParser()
    parser.read(ini_filename)
    conn_info = {param[0]: param[1] for param in parser.items("postgresql")}
    return conn_info

def insert_data(query: str, conn: psycopg2.extensions.connection, cur: psycopg2.extensions.cursor, df: pd.DataFrame, page_size: int) -> None:
    data_tuples = [tuple(row.to_numpy()) for index, row in df.iterrows()]

    try:
        psql_extras.execute_values(
            cur, query, data_tuples, page_size=page_size)
        print("Запрос:", cur.query)

    except Exception as error:
        print(f"{type(error).__name__}: {error}")
        print("Запрос:", cur.query)
        conn.rollback()
        cur.close()

    else:
        conn.commit()
        
def get_column_names(table: str, cur: psycopg2.extensions.cursor) -> List[str]:
        cursor.execute(f"SELECT column_name FROM INFORMATION_SCHEMA.COLUMNS WHERE TABLE_NAME = '{table}';")
        col_names = [result[0] for result in cursor.fetchall()]
        return col_names

def get_data_from_db(query: str, conn: psycopg2.extensions.connection, cur: psycopg2.extensions.cursor, df: pd.DataFrame, col_names: List[str]) -> pd.DataFrame:
    try:
        cur.execute(query)
        while True:
            # Разберем следующие 100 строк таблицы
            query_results = cur.fetchmany(100)
            # Если возвращается пустой список, значит, мы достигли конца результатов
            if query_results == list():
                break

            # Создаем список словарей, в котором каждый словарь представляет собой одну строку
            results_mapped = [
                {col_names[i]: row[i] for i in range(len(col_names))}
                for row in query_results
            ]

            # Добавьте извлеченные строки во фрейм данных
            df = df.append(results_mapped, ignore_index=True)

        return df

    except Exception as error:
        print(f"{type(error).__name__}: {error}")
        print("Query:", cur.query)
        conn.rollback()    


cols = ["X_axis" ,"y_axis","z_axis" ,"pressure" ,"сapture_angle" ,"time_label" ,"test_id"]
df = pd.read_table('test_image.txt', names = cols, sep=";")
#print(df)


conn_info = load_connection_info("db.ini")

connection = psycopg2.connect(**conn_info)
cursor = connection.cursor()
col_names = get_column_names('image', cursor)

house_df = pd.DataFrame(columns=col_names)
query = "SELECT * from homework.image;"
test_df = get_data_from_db(query, connection, cursor, df, col_names)
print(test_df)

connection.close()
cursor.close()


```
